## IMPORT LIBRARY

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
```

## MEMBACA GAMBAR

```python
img = cv2.imread('coloring.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

## UBAH KE HSV

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

## AMBANG BATAS BIRU,MERAH,HUJAU

```python
lower_blue = np.array([100, 150, 0])
upper_blue = np.array([140, 255, 255])
mask_blue = cv2.inRange(img_hsv, lower_blue, upper_blue)

lower_red1 = np.array([0, 120, 70])
upper_red1 = np.array([10, 255, 255])
mask_red1 = cv2.inRange(img_hsv, lower_red1, upper_red1)

lower_red2 = np.array([170, 120, 70])
upper_red2 = np.array([180, 255, 255])
mask_red2 = cv2.inRange(img_hsv, lower_red2, upper_red2)

mask_red = mask_red1 | mask_red2
mask_red_blue = mask_red | mask_blue

lower_green = np.array([40, 40, 40])
upper_green = np.array([90, 255, 255])
mask_green = cv2.inRange(img_hsv, lower_green, upper_green)

mask_rgb = mask_red | mask_green | mask_blue
```

## MENAMPILKAN HASIL

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(20, 10))
ax1.imshow(np.zeros_like(mask_blue), cmap='gray')
ax1.set_title("NONE")
ax1.axis('off')

ax2.imshow(mask_blue, cmap='gray')
ax2.set_title("BLUE")
ax2.axis('off')

ax3.imshow(mask_red_blue, cmap='gray')
ax3.set_title("RED-BLUE")
ax3.axis('off')

ax4.imshow(mask_rgb, cmap='gray')
ax4.set_title("RED-GREEN-BLUE")
ax4.axis('off')

plt.tight_layout()
plt.show()
```
