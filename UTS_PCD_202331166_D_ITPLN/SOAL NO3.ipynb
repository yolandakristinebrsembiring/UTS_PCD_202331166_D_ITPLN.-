## IMPORT LIBRARY

```python
from PIL import Image, ImageEnhance
import matplotlib.pyplot as plt
import cv2
import numpy as np
import os
```

## MEMBACA GAMBAR dan MEMBUKA GAMBAR ASLI

```python
image_path = "backlight.jpg"
```

```python
if not os.path.exists(image_path):
    print(f"‚ùå File tidak ditemukan: {image_path}")
else:
    original = Image.open(image_path)
```

## KONVERSI KE GRAYSCALE

```python
gray = original.convert("L")
```

## Mencerahkan grayscale

```python
gray_np = np.array(gray)
bright_gray_np = cv2.convertScaleAbs(gray_np, alpha=1.0, beta=50)  # beta: tambah kecerahan
bright_gray = Image.fromarray(bright_gray_np)
```

## MENINGKATKAN KONTRAS GRASYSCALE

```python
 contrast_enhancer = ImageEnhance.Contrast(gray)
 contrast_gray = contrast_enhancer.enhance(2.0)
```

## GABUNGKAN CERAH DENGAN KONTRAS

```python
bright_then_contrast_np = cv2.convertScaleAbs(gray_np, alpha=2.0, beta=50)
bright_then_contrast = Image.fromarray(bright_then_contrast_np)
```

## MENAMPILKIAN HASIL

```python
fig, axes = plt.subplots(1, 5, figsize=(20, 5))

axes[0].imshow(original)
axes[0].set_title("Gambar Asli")

axes[1].imshow(gray, cmap='gray')
axes[1].set_title("Grayscale")

axes[2].imshow(bright_gray, cmap='gray')
axes[2].set_title("Grayscale Dicerahkan")

axes[3].imshow(contrast_gray, cmap='gray')
axes[3].set_title("Grayscale Dikontras")

axes[4].imshow(bright_then_contrast, cmap='gray')
axes[4].set_title("Cerah + Kontras")
```
